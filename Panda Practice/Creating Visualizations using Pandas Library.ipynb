```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data
# set date column as the index column
df = pd.read_csv(r"C:\Users\wjlee\Desktop\Data Analytics Learning\Sample Data Sets\Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
# this is to show the available chart style from matplotlib. if you dont define what kind of chart style to use, all the charts will be of a default style
# plt.style.use will adopt the style
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
# df.plot is to visualize (meaning we run plot on df (hence df.plot) default plot = line (kind = 'line')
# shift + tab to see more info. under kind, can specify what kind of visualization do we want
df.plot()
```

```python
# in the above query, all data is combined into one graph. subplots = true would then break the data into its each individual graph, as below
df.plot(kind = 'line', subplots = True)
```

```python
# this will add a title for the graph. xlabel and ylabel will provide a label for x and y axis
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
# use kind = 'bar' to change graph type to bar graph. use stacked = True to create a stacked bar chart (as per below query)
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
# this returns only the values for flavor ratings. note to include the definition after df, not after plot. 
df['Flavor Rating'].plot(kind = 'bar')
```

```python
# kind = 'barh' means horizontal bar chart. 
df['Flavor Rating'].plot(kind = 'barh')
```

```python
# unstacked
df.plot(kind = 'barh')
```

```python
# stacked view
df.plot(kind = 'barh', stacked = True)
```

```python
# this query wouldnt work because we need to specify x and y axis (last error output: ValueError: scatter requires an x and y column)
df.plot(kind = 'scatter')
```

```python
# can either define chart type as kind = 'scatter', or alternatively, just define it straight after plot (df.plot.scatter). see below examples
df.plot(kind = 'scatter', x = 'Texture Rating', y = 'Overall Rating')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.bar(stacked = True)
```
