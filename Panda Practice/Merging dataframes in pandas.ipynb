```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\wjlee\Desktop\Data Analytics Learning\Sample Data Sets\Pandas\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\wjlee\Desktop\Data Analytics Learning\Sample Data Sets\Pandas\LOTR 2.csv")
df2
```

```python
# by default, the first DF will be the left DF
# by default, it will be inner join. but this query does not state join on which col
df1.merge(df2)
```

```python
# in this case, how = 'inner' means we define the join as inner join. 
# we also define the join columns on 'FellowshipID'
# you will notice that the output is FirstName_x and FirstName_y. this is because the join col is on fellowshipID, and will return cols on both tables (except for the joint col)
df1.merge(df2, how = 'inner', on = 'FellowshipID')
```

```python
# doing it this way (where 'FellowshipID' and 'FirstName' are placed in list [], output will be same as before (no repeat of firstname col)
# this is because now we specify that the join is on both the common cols, which are 'FellowshipID' and 'FirstName'
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
# this will return all values, regardless if they are the same. hence, some fields are NaN (null)
df1.merge(df2, how = 'outer')
```

```python
# all data from left table (df1), and whatever joint values from right table (df2)
df1.merge(df2, how = 'left')
```

```python
# all data from right table (df2), and whatever joint values from left table (df1)
df1.merge(df2, how = 'right')
```

```python
# take each value from left df (x) and compares with each value in right df (y)
df1.merge(df2, how = 'cross')
```

```python
# wont work because merge had a lot of default functions, but join need to define the cols to join
df1.join(df2)
```

```python
# in this case, we are doing outer join. hence, since it's a full join on both tables, need to specify what cols to join on 'FellowshipID'
# also need to specify suffix, which is the col_name+which table it belongs to (left table or right table). the _Left/_Right is to add on after ColName to indicate that this col is from which table
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left', rsuffix = '_Right')
```

```python
# joins usually work better with INDEX (first col). for merge, usually it's on col names, and it's relatively easy to do
```

```python
# in this case, we are creating a new df (df4), and we join the FellowshipID as the index column
# output will be inner join(in pd, is merge), but we define the first_name col (since both table has that col) and define which table it came from
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right')
df4
```

```python
# same as above, but we set the df, and this is now an outer join
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

```python
# concatenate is like putting a df on top of another. merge/join is putting the df next to each other
# you will see that for join, it is like combining the values into the same row through a common column. but for concat, it's basically copy and paste the tables together
# hence, fellowship id has 1001 and 1002 and it's repeating the same values again, since both 1001 and 1002 appears in both df
# also, you will see that it is not combining the values tgt. this means that df1 does not have age col. hence, 1001 age is NaN (cuz it's from df1)
# while at the bottom, 1001 has age but no skills, because that data is from df2
pd.concat([df1,df2])
```

```python
# in this case, it is still concat, means it's laying df2 on top of df1. however, since we specified the join to be inner, this means that the output will only show what is common col between both dfs, which is fellowshipID and firstname
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
# in this case, it's outer join, but we "grouped" the axis together by index (in the above line, the index are 0,1,2,3,4), so the output now can see that 0,1,2,3,4 are merged together
# meaning in the above reference, index 0/fellowshipID 1001 has hiding skills (df1) and is 39 age (df2)
# but when merged, will see that index 0/fellowshipid 1001 is now a single line value that combines both dfs 
pd.concat([df1,df2], join = 'outer', axis = 1)
```
