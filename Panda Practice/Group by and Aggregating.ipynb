```python
# in the above example, difficult to group flavor columns as they are all unique values. better to group those with duplicate or similar values. 
# base flavor would be good to group
# in the below example, once we group it, need to define it with a name (group_by_frame)
df.groupby('Base Flavor')
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame
```

```python
# in this case, we are trying to find the average values of the base flavor group. need to specify numeric_only=True. 
# if never specify, then means that may return as error because there may be data that is not in int, so if in str or other data type, may not be able to perform aggregate
group_by_frame.mean(numeric_only=True)
```

```python
# this is just the standard way of writing, if it's not in the defined name of group_by_frame
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
# count
df.groupby('Base Flavor').count()
```

```python
# min. in this case, it will group by the base flavor, but because it is min, it will sort the minimum value within the base flavor group
# this means that in the chocolate base flavor group, the minimum value is chocolate, and in the vanilla base flavor group, the min value is cake batter
df.groupby('Base Flavor').min()
```

```python
# max. in this case, it will group by the base flavor, but because it is min, it will sort the max value within the base flavor group
# this means that in the chocolate base flavor group, the max value is rocky road, and in the vanilla base flavor group, the max value is vanilla
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
# in the above example, need to include numeric_only=True, because otherwise, everything will be added up
df.groupby('Base Flavor').sum(numeric_only=True)
```

```python
# in this case, we are creating a new table with aggregated values. need to define aggregate what values (in this case, it will be flavor rating)
# for the aggregated values, create a list [] and to list out the functions (mean, max, count, sum)
df.groupby('Base Flavor').agg({'Flavor Rating':['mean', 'max', 'count', 'sum']})
```

```python
# in this case, to create 2x aggregate value
df.groupby('Base Flavor').agg({'Flavor Rating':['mean', 'max', 'count', 'sum'],'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

```python
df
```

```python
# in this case, we group 2 columns at one go
# output below shows that for chocolate, only had yes for liked. but for vanilla group, got some say yes, some say no
df.groupby(['Base Flavor', 'Liked']).mean(numeric_only=True)
```

```python
# in this case, we grouped by 2 column (base flavor and liked), and queried the results to show the mean, max, count, sum for flavor rating
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating':['mean', 'max', 'count', 'sum']})
```

```python
# describe = shows high level overview for each column, broken down to std (standard deviation?), min, 25%, 50%, 75%, count, mean
df.groupby('Base Flavor').describe()
```
