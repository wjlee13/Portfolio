```python
# in this case, can use strip to remove unwanted characters. for removal of unwanted char at the front, use lstrip. for the back, use rstrip
# need to specify data type also. in this case, it's string, so str
# also, need to define columns. if never define columns, the code will apply to the whole df (since the code is df.str.strip)
# hence, to specify column, use df["col_name"] (make sure there's no period or full stop after df. this is because full stop means the start the next part of the query/instruction)
# strip() the default will be to remove spacing
df["Last_Name"].str.strip()
```

```python
# remove '...' from potter
df["Last_Name"].str.lstrip("...")
```

```python
# remember to save to df.
# in this case, PLEASE REMEMBER NOT TO SAVE AS df = df["Last_Name"].str.lstrip("..."). 
# this is because if written that way, will only save the column as df
# save it as df["Last_Name"]. this means that we are only making changes to the "Last_Name" col, and saving the changed Last_Name col into the existing df. everything else stays the same
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df
```

```python
# this removes the "/" for 1003 white and 1015 Flenderson
df["Last_Name"] = df["Last_Name"].str.lstrip("/")
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df
```

```python
# the above used lstrip and rstrip
# an alternative would be to just use strip and insert the characters to be stripped, and it will apply to all
# use this df["Last_Name"].str.strip("123._/")  [no need to put characters into a list, can put them all in just one value]
```

```python
# this is to replace characters in the Phone_Number column. when use replace, need to define what character to remove, and also what character will be replacing the removed character
# in this case, we are saying ^ means NOT. we are also saying a-z, A-Z, 0-9, which means that any character that is not a-z, A-Z, 0-9, to replace with space ('')
# also remember to include regex = True. this means regular expression syntax
df["Phone_Number"].str.replace('[^a-z,A-Z,0-9]','',regex = True)
```

```python
# this is to show if we do not include ^, it means that all a-z, A-Z, 0-9 will be replaced by space''. hence, remember to include ^
df["Phone_Number"].str.replace('[a-z,A-Z,0-9]','',regex = True)
```

```python
df["Phone_Number"].str.replace('[^a-z,A-Z,0-9]','',regex = True)
```

```python
# saving the above queries into df. remember, save the column, else will just overwrite and save the whole df
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-z,A-Z,0-9]','',regex = True)
df
```

```python
# lambda is anonymous functions that can be used to perform concise operations on DataFrames and Series. kind of like an one-off open function.
# in this case, we use lambda x to change data type for Phone_Number from string to float
# instead of applying lambda function to convert Phone_Number column elements to string , we can also use 
# df['Phone_Number'] = df['Phone_Number'].astype(str)
df["Phone_Number"].apply(lambda x: str(x))
```

```python
# save as df
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
# now that phone number is converted from str to float, let's use lambda to separate the values and place - in between xxx-xxx-xxxx
# 0:3 because in python, it starts with 0. it goes up to position 3, so need to say 3 instead of 2 (think of it as having to complete the entirety of 2)
# 3:6 start with 3 cuz inclusive of position 3
df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
# save as df
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# in this case, we are working on the address column, and we are splitting that column. the delimiter is comma, so we split at the commat ','
# next, we see n=2. n means column. this means that we are splitting the values into 3 columns (in python, starts with 0, so 0,1,2 (3 columns in total)
# expand = true would place it in table format
df['Address'].str.split(',', n=2, expand = True)
```

```python
# save as df
# remember, cannot save over it, especially now that a column is split into 3 columns, so when saving, need to save the query into 3 columns and name them accordingl "Street_Address, State, Zip Code"
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand = True)
df
```

```python
# in this case, we specify x as a for loop, whereby the code goes through each row to check
# if is the conditional statement, where by we go through do not contact column, if it's Y, then drop it from the df
# The inplace=True argument modifies the DataFrame directly, rather than creating a new copy.
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)

df
```
